```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences = [
    'I love programming in Python',
    'Python is a great programming language',
    'I enjoy learning new programming languages',
    'Machine learning is fascinating',
    'Deep learning is a subset of machine learning'

]
```

```python
# Define the vocabulary size
vocubalary_size = 10000
```

```python
# One hot representation of the sentences
one_hot_representations = [one_hot(words, vocubalary_size) for words in sentences]
one_hot_representations
```

```python
# Word Embedding representation
from tensorflow.keras.layers import Embedding
# from tensorflow.keras.preprocessing.sequence import pad_sequences

from tensorflow.keras.utils import pad_sequences

from tensorflow.keras.models import Sequential
```

```python
import numpy as np  
```

```python
max_sentence_length = 10
embedded_docs = pad_sequences(one_hot_representations, maxlen=max_sentence_length, padding='pre')
print(embedded_docs)
```

```python
# Feature representation 
dimension = 10
```

```python
model = Sequential()
model.add(Embedding(vocubalary_size, dimension, input_length=max_sentence_length))
model.build(input_shape=(None, max_sentence_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
